```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Embedding, Bidirectional, SimpleRNN, Dense, LSTM, GRU
```

```python
# Load the IMDb dataset
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

# Pad sequences to have the same length
x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100)
```

```python
# Define the RNN model
model_simple_rnns = Sequential([
    Input(shape=(100,)),
    Embedding(10000, 32), # Embedding layer to convert words to vectors
    SimpleRNN(5, return_sequences=True),    # RNN layer with 5 units
    SimpleRNN(5),                           # Another RNN layer with 5 units
    Dense(1, activation='sigmoid')          # Output layer for binary classification
])

model_simple_rnns.summary()
```

```python
# Define the LSTM model
model_stacked_lstms = Sequential([
    Input(shape=(100,)),
    Embedding(10000, 32,),
    LSTM(5, return_sequences=True),
    LSTM(5),
    Dense(1, activation='sigmoid')
])

model_stacked_lstms.summary()
```

```python
# Define the GRU model
model_stacked_grus = Sequential([
    Input(shape=(100,)),
    Embedding(10000, 32,),
    GRU(5, return_sequences=True),
    GRU(5),
    Dense(1, activation='sigmoid')
])

model_stacked_grus.summary()
```

```python
# Compile the model
model_simple_rnns.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_stacked_lstms.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_stacked_grus.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history_simple_rnns = model_simple_rnns.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
# Train the model
history_stacked_lstms = model_stacked_lstms.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
# Train the model
history_stacked_grus = model_simple_rnns.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
# Define the GRU model
model_bi_rnn = Sequential([
    Input(shape=(100,)),
    Embedding(10000, 32),
    Bidirectional(SimpleRNN(5)),
    Dense(1, activation='sigmoid')
])

model_bi_rnn.summary()
```

```python
# Define the GRU model
model_bi_lstm = Sequential([
    Input(shape=(100,)),
    Embedding(10000, 32),
    Bidirectional(LSTM(5)),
    Dense(1, activation='sigmoid')
])

model_bi_lstm.summary()
```

```python
# Define the GRU model
model_bi_gru = Sequential([
    Input(shape=(100,)),
    Embedding(10000, 32,),
    Bidirectional(GRU(5)),
    Dense(1, activation='sigmoid')
])

model_bi_gru.summary()
```

```python
# Compile the model
model_bi_rnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_bi_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_bi_gru.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history_bi_rnn = model_bi_rnn.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```
